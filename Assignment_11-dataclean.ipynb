# ASSIGNMENT 11- Data Cleaning



```python
import pandas as pd
import numpy as np
#Creation of the given dataframe
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
#Printing the dataframe
df
```

```python
#Q.1 Replacing the NaN values with the elements fitting the mentioned pattern
for i in range(df.shape[0]):
    if np.isnan(df.loc[i, 'FlightNumber']):
        df.loc[i, 'FlightNumber'] = df.loc[i-1,'FlightNumber']+10
    else:
        pass
#Converting the float column as an integer column
df['FlightNumber'] =df['FlightNumber'].astype(int)
df
```

```python
#Q.2
#Creation of a temporary dataframe
temp_df = pd.DataFrame()
#Populating the dataframe with the split values of  the column
temp_df['From'] = df['From_To'].str.split('_').str[0]
temp_df['To'] = df['From_To'].str.split('_').str[1]
#printing the temporary dataframe
temp_df
```

```python
# Q.3
#Standardising the strings  in the temporary dataframe so that only the first letter is uppercase 
temp_df = temp_df.apply(lambda x: x.astype(str).str.capitalize())
#Printing the standardised dataframe
temp_df
```

```python
#Q.4
#Deleting the From_To column from the database
del df['From_To']
#Attaching the temporary dataframe to the original one
df = pd.concat([temp_df,df],axis=1)
df
```

```python
#Q.5
#Expanding the Series of lists into a DataFrame named delays
delays = pd.DataFrame(df.pop('RecentDelays').values.tolist())
#Renaming the columns delay_1,delay_2 
lst=list(delays.columns)
for i in range(delays.shape[1]):
    lst[i]='delay_'+ str(i+1)
delays.columns = lst
delays
```

```python
#Adding delays to the original dataframe
df = pd.concat([df,delays],axis=1)
#Printing the modified dataframe
df
```
